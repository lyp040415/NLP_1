```python
import nltk
from nltk.corpus import gutenberg
nltk.download('gutenberg')
from collections import Counter
import math
```

```python
# 加载Gutenberg语料库中的文本
text = gutenberg.raw("shakespeare-hamlet.txt")  # 原始文本
words = gutenberg.words("shakespeare-hamlet.txt")  # 分词后的文本

# 计算字母级信息熵
letters = [char.lower() for char in text if char.isalpha()]  # 提取字母并转为小写
letter_freq = Counter(letters)  # 统计字母频率
total_letters = len(letters)  # 总字母数
letter_entropy = 0.0

for letter, count in letter_freq.items():
    prob = count / total_letters  # 计算概率
    letter_entropy -= prob * math.log2(prob)  # 累加熵值

print(f"字母级信息熵: {letter_entropy:.4f}")

# 计算单词级信息熵
words = [word.lower() for word in words if word.isalpha()]  # 提取纯单词并转为小写
word_freq = Counter(words)  # 统计单词频率
total_words = len(words)  # 总单词数
word_entropy = 0.0

for word, count in word_freq.items():
    prob = count / total_words  # 计算概率
    word_entropy -= prob * math.log2(prob)  # 累加熵值

print(f"单词级信息熵: {word_entropy:.4f}")
```

```python
import matplotlib.pyplot as plt

letter_freq = sorted(letter_freq.values(), reverse=True)  # 按频率排序

plt.figure(figsize=(10, 6), dpi=500)
# 绘制中文词频长尾图
plt.plot(range(1, len(letter_freq) + 1), letter_freq, label="English Word Frequency", color="red")

# 设置对数坐标轴
plt.xscale("log")
plt.yscale("log")

# 添加标题和标签
plt.title("Long-tail Distribution of Letter Frequencies (English)")
plt.xlabel("Letter Rank (log scale)")
plt.ylabel("Letter Frequency (log scale)")
plt.legend()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

word_freq = sorted(word_freq.values(), reverse=True)  # 按频率排序

plt.figure(figsize=(10, 6), dpi=500)
# 绘制中文词频长尾图
plt.plot(range(1, len(word_freq) + 1), word_freq, label="English Word Frequency", color="red")

# 设置对数坐标轴
plt.xscale("log")
plt.yscale("log")

# 添加标题和标签
plt.title("Long-tail Distribution of Word Frequencies (English)")
plt.xlabel("Word Rank (log scale)")
plt.ylabel("Word Frequency (log scale)")
plt.legend()

# 显示图形
plt.show()
```
